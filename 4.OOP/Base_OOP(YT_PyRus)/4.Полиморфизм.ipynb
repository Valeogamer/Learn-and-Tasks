>  Полиморфизм - это механизм, позволящий выполнять один и тот же код по разному. 

>Ducktyping (утиная типизация) - наличие поведения(методов) для использования в полиморфизме (если что то ходит и крякает как утка то мы считаем что это утка,
независимо от того является ли это уткой на самом деле).

> В ЯП со статическо типизацией для полиморфизма важно кто ты (какой тип), для питона важно что ты умеешь. А для джавы допустим важно какой тип, значит оно примет только с одинаковым типом, а для питона важно только умение. То есть имена методов одинаковые, а для джавы чтобы сам тип класса был одинаков.

то есть мы хотим функцию с одним именем, но допустим с использованием различных типов данных.

```python
class Animal:
  def make_noise(self):
    print('shh')

class Cat(Animal):
  def make_noise(self):
    print('Meow')

class Dog(Animal):
  def make_noise(self):
    print('gavv')

class Car: # не наследуем мы же помним IS A является, машина не Животное
  def make_noise(self):
    print('bi-bi')

# данная запись указывает какой конкретный тип мы ждем animal:Animal -> 
# -> имя атрибута : имя типа, только ждет и если мы передадим что то другое 
# пример с авто, оно будет работать, но только в PYTHON, на других языках данный
# трюк не прокатит

def noise(animal:Animal):
  animal.make_noise()

if __name__ == '__main__':
  noise(Dog())
  noise(Cat())
  noise(Car()) # о чудо работает -> утиная типизация, а сработало потому что 
  # наша машина умеет что то схожее делать, но это совсем другой объект
  # это неправильно, но будет работать, так как имя методов одинаково
```

```python
class SQLiteDatabase:
  def connect(self):
    print("Connectin")

  def get_users(self):
    print('get users with SQL')

class MongoDB:
  def connect(self):
    print("Connectin")

  def get_users(self):
    print('get users with NoSQL')

class Server:
  def get_users(self, db):
    db.connect()
    users = db.get_users()
    return users


if __name__ == '__main__':
  server = Server()
  server.get_users(MongoDB()) # менять базу данных туть
```
