В python все является объектами

> __*Объект__ - сущность, объединяющая данные и методы для работы с ними (__*состояние*__ (данные объекта) = {что знает} и __*поведение*__ (методы)) = {что умеет}

> Чертеж (это класс), а объект уже конкретный дом.

> __**Класс*__ - это новый тип данных, объект его конкретный представитель.

Каждый раз когда мы создаем класс, мы создаем новый тип данных, допустим точно такой же как int, float, и остальные.

> у любого объекта есть id() и type() и значение

> По PEP8 по определенной конвенции, в классе заглавная буква большая

> 1. Первая потребность классов, когда не хватает встроенных типов или функционала.
2. Разные состояния.


> __*Метод__ - это функция, которая привязана к классу, принадлжеит классу, поэтому его называют метод.

> __*self__ - адрес объекта, то есть вызовет не метод класса или атрибут(свойство), а именно по адресу памяти, метод который у объекта или атрибут.\
Атрибут _*self*_ подставляется python неявно \
__*self__ - это ссылка на конкретный объект

> __*магические методы__  или дандер методы(дабл андер двойной поджопник) или волшебные методы

> В методах пишем через self, потому что мы знаем что простые переменные не живут вне функции, если с ними работать только там, а с помощью self, мы обращаемся по адресу паматя к нужным нам атрибутам.

>   
    # self ссылка на конкретный объект
    # конкретному объекту присваиваем атриубуты
    # грубо говоря мы обращаясь по данному адресу к манекену шаблону объекта, 
    # добавляем некоторые атрибуты, и по адресу мы точно знаем что определенный
    # атрибут лежит именно там по такому то адресу.

```python
from ast import pattern
class Cat:
  def __init__(self, name, age): # конструктор (инициализатор)
    self.name = name  # атрибут для конкретного объекта
    self.age = age # мы конкретного 
    
  def meow(self): # первый аргумент во всех методах будет self
    print(self)  # == print(Tom)
    print(f'{self.name} says: Meow!') # 

if __name__ == "__main__":
  tom = Cat('Tom', 2)  # а вот объект том это уже конкретный тип данный, представитель
  angela = Cat('Angela', 1) # а вот это уже другой тип объекта 
  # print(tom is angela) # а вот и проверка этого 
  # print(tom == angela) # та же проверка как и в is по ссылке в памяти
  print(tom) # без репра выводит просто адреса памяти
  print(angela)
  # print(type(tom)) # тип объекта
  # print(type(angela))
  # print(id(tom))
  # print(id(angela))
  tom.meow()  # неявно через self, передачей адреса памяти
  # Cat.meow(tom) #!!! явно передали
  angela.meow()
```
