> LEGB - rule действует для атрибутов без приставок (self)

> для self атрибутов поиск идет сначала в объекте, потом в классе, затем у предков OCP(object-class-parents) - это все как пайтое ищет такие атрибуты через self. Как видим сперва в области видимости у объека(экземпляра класса), если у экземпляра класса не находит, то далее ищет у класса, и если и у класса не находит, то обращается к предкам и пытаемся найти в области видимости предка.

> Если через self пытаться поменять неизменяемые атрибут (строка) пример с breed, то будет создана локальная копия. Но если будет изменяемой атрибут, то он измениться для всех объектов класса.

> Через self атрибуты класса менять не желательно!

> @classmethod нужен для того чтобы работать с переменными с атрибутами класса и атрибутами класса

> cls - это ссылка на сам класс

> @classmethod используется для работы с атрибутами класса и с другими методами класса и только с ними! И они будут помечены декоратором @classmethod и они видят только вот методы с таким декоратором!!!

> @staticmethods - не получает ссылок под капотом, это просто функция связанная контекстом с классом

> Если у нас есть метод в котором self или cls не используются, то это статик метод, PEP8 и Pycharm подскажет, что это возможно статик.

```python
class BlueCat:
  breed = 'Russian Blue' # атрибуты класса
  names = []
  count = 0

  def __init__(self, name: str, age: float):
    self.name = name # атрибут объекта
    self.age = age # атрибут объекта
    # BlueCat.breed = 'other' # только так поменется для всех
    # таким образом меняем атрибут класса для всех
    self.names.append(name) # если мы будем менять атрибут класса, который является
    # изменяемым то он изменится для всех
    # self.inc_count() # обновление счетчика
    # то есть при создание любого экземпляра мы автоматически прибавим +1 
    # вызвав метод класса, который работает с атрибутами класса
    # правильнее даже будет вот так
    BlueCat.inc_count()

  def meow(self):
    print(f'{self.name} of {self.breed} says: Meow!')
  
  @classmethod
  def inc_count(cls):  # сюда приходит именно сам класс BlueCat, а не экземпляры с его атрибутами
    cls.count += 1
  
  @classmethod
  def make_cat(cls, name): # некий конструктор, чтобы конструировать объекты класса
    if name == 'Tom':
      return cls('Tom', 2) # cls == BlueCat, разницы никакой
    elif name == 'Angela':
      return BlueCat('Angela', 1.5)
    else:
      return cls('Ginger', 0.5)
  
  @staticmethod
  def get_human_age(age): # чтобы было очевидно для чего данный и метод и вместе с чем 
    return age * 8  # его применять
  
  def get_h_age(self, age): # PyCharm скажет что возможно это статик, так как
    return age * 8 # мы этот self никак не используем


if __name__ == '__main__':
  # Tom = BlueCat('Tom', 2) # смотри ниже альтернативное создание 
  Tom = BlueCat.make_cat('Tom') # чтобы передав 1 атрибут, остальные атрибуты 
  # передавались по дефолту
  # Angela = BlueCat('Angela', 2)
  Angela = BlueCat.make_cat('Angela')
  Tom.breed = 'other' # Поменяли атрибут объекта Том, но перед этим создали у 
  # у тома такое свойство, атрибут и потом присвоили, то есть не стали искать в 
  # в области видимости атрибута класса
  Tom.meow()
  Angela.meow()
  print(Tom.breed)
  print(Tom.names)
  print(Angela.names)
  print(BlueCat.count) # вывод значения счетчика
  print(BlueCat.get_human_age(Angela.age))
```
