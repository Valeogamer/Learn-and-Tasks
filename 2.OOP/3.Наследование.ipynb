### IS - A является (наследование)
### HAS - A содержит (композиция)

двойное подчеркивание, чтобы случайно не затерли значение нашего предка, от наследника мы получили и у самого себя и у отца, это все позволяет сделать private \
Двойное подчеркивание оно сделано для того чтобы при наследовании случайно не переопределили некоторую секретную переменную своего предка. \
Оно не для закрытия данных, защита от переопределения!

```python
# Inheritance - Наследование, это механизм получения доступа к данным и поведение своего предка.
# Первое лицо наследования механизм получения доступа к данным и поведениям своего предка
# А второе расширение (изменение поведения) классов не меняя код
# Правило наследования IS - A - является, правильное наследование должно выполнять вот этому правилу
# Тем самым говорим что наш наследник является наследуемым вот от этого вот
# Директор -> Работник, но нельзя наследовать от Директора уборщицу или наоборот
# Уборщица -> Работник
# Менеджер -> Работник
# не стоит забывать то что от кого наследуемся мы получаем всего его методы и поведения и так далее
# то есть если директор наследуется от уборщицы то получит способность мыть полы и протирать пыль

class Employee:
    def __init__(self, name, salary, bonus):  # конструктор
        self.name = name
        self.salary = salary
        self.bonus = bonus

    def calc_total_bonus(self):
        return self.salary // 100 * self.bonus

    def __str__(self):
        return f'{self.__class__.__name__} {self.name}, salary = {self.salary}, bonus={self.bonus}%, ' \
               f'total bonus={self.calc_total_bonus()} rub'


class Cleaner(Employee):
    def __init__(self, name):  # конструктор
        super().__init__(name, 15000, 1)  # через супер получили доступ к конструктуру предка


class Manager(Employee):
    def __init__(self, name):  # конструктор
        super().__init__(name, 45000, 15)


class CEO(Employee):
    def __init__(self, name):  # конструктор
        super().__init__(name, 105000, 100)

    """
    Если захочу изменит какой то метод то или написать вообще свой метод,
    то создать свой класс и метод
    Переопределять поведение - переписать метод своего предка
    """

    def calc_total_bonus(self):  # так же Pycharm спец значком показывает что мы переопределили данный метод
        return 200_000

class MyList(list):
    def __str__(self):
        return super().__str__().replace(',', ',\n')

if __name__ == '__main__':
    masha = Cleaner('Mar Ivan')
    print(masha)
    print(dir(masha))  # dir - посмотреть свойства
    print(masha)
    grisha = Manager('Grisha Pe')
    print(grisha)
    ivan = CEO('Ivan Pavel')
    print(ivan)
```

```python
"""
Создали некий класс, унаследовались от того, от которого хотели получить поведение
и переопределили его метод __STR__
"""
from pprint import pprint

class MyList(list):  # Наследовались от list
    def __str__(self):
        return super().__str__().replace(',', ',\n')  # super способ обращения к методам предка, верни нам свой
    # стандартный __str__ и меняем каждую запятую с переносом строки, переопределили str стандартный


if __name__ == '__main__':
    print([1, 2, 3])
    my_list = MyList([1, 2, 3])
    print(my_list)
    print(my_list[1])  # чтение по индексу он взял от родителя
    my_list.extend([4, 5]) # от родителя также взял extend
    print(my_list)
    pprint(dir(my_list))
```

```python
class Cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def meow(self):
        print(f'{self.name} says: Meow!')

    def __add__(self, other):  # позволяет складывать
        if isinstance(other, Cat):
            return Cat('Ginger', 0)


if __name__ == '__main__':
    tom = Cat('Tom', 2)
    angela = Cat('Angela', 1)
    ginger = tom + angela
    ginger.meow()
```
